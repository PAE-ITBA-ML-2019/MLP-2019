```python
from mlp_helper import generate_gaussians_distributions, plot_boundaries, draw_neural_net, return_weights_notation
from matplotlib import pyplot as plt
import numpy as np
```

```python
X, y = generate_gaussians_distributions(sep=1, N = 50, normalize=False, random_state=41)
```

```python
plot_boundaries(X, y)
```

```python
fig, ax = plt.subplots(1,1,figsize=(8, 8))
shape = [2,2,1]
weights, biases = return_weights_notation(shape)
draw_neural_net(ax, 0.1, 0.9, 0, 0.9, shape, weights, biases)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(2, input_shape=(2,), activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
from keras.optimizers import Adam
model.compile(Adam(lr=0.01), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=1000, verbose=0, batch_size=128)
```
